# Problem 1 Aradaki Tam Sayılar

Kullanıcıdan iki adet sayı alınız, bu sayılar arasındaki tam sayıları ekrana yazdıran bir program yazınız. 

```python
baslangic = int(input("Başlangıç sayısını girin: "))
bitis = int(input("Bitiş sayısını girin: "))

if baslangic > bitis:
    baslangic, bitis = bitis, baslangic  # sayıları küçükten büyüğe sırala

for sayi in range(baslangic, bitis+1):
    print(sayi)
```

# Problem 2 Faktöriyel

Kullanıcıdan bir sayı alınız ve alınan sayının faktöriyelini hesaplayan bir program yazınız. 


```python
sayi = int(input("Faktöriyeli hesaplanacak sayı: "))
faktoriyel = 1

for i in range(1, sayi+1):
    faktoriyel *= i

print("{}! = {}".format(sayi, faktoriyel))
```

# Problem 3 Virgül ile Ayırma

Kullanıcıdan bir metin halinde virgül ile ayrılmış sayılar alınız ve aldığınız her bir sayıyı listeye atayınız, daha sonra listedeki her elemanı toplayıp ekrana bastıran bir program yazınız.

<h4>Sample Input</h4>
<p>
Lutfen sayilari virgul ile ayrirarak giriniz: 5,14,7,2,1
</p>
<h4>Output</h4>

Toplam: 29

```python
sayilar = input("Lutfen sayilari virgul ile ayrirarak giriniz: ")
sayi_listesi = sayilar.split(",")
toplam = 0

for sayi in sayi_listesi:
    toplam += int(sayi)

print("Toplam:", toplam)
```

```python
import datetime
dogum_tarihi = input("Doğum tarihinizi (GG/AA/YYYY) girin: ")
dogum_gunu, dogum_ayi, dogum_yili = map(int, dogum_tarihi.split('/'))
bugun = datetime.date.today()
dogumgunu_bu_yil = datetime.date(bugun.year, dogum_ayi, dogum_gunu)
yas = bugun.year - dogum_yili - ((bugun.month, bugun.day) < (dogum_ayi, dogum_gunu))
yasam_suresi = bugun - dogumgunu_bu_yil
print("Bugüne kadar yaşadığınız süre: {} yıl, {} ay ve {} gün".format(yas, yasam_suresi.month, yasam_suresi.day))
```

 # Problem 10 Altıgen Çizimi

Kullanıcıdan altıgenin kaç satırdan oluşacağı ve orta satırında kaç karakter olacağı bilgisini alarak '*' karakteri ile bir altıgen çizin. (For loop kullanabilirsiniz.)

```python
n = int(input("Altıgenin kaç satırdan oluşacağını girin: "))
middle_row = int(input("Orta satırda kaç karakter olacak?: "))

if middle_row > 2*n-1:
    print("Hatalı giriş! Orta satırda daha fazla karakter olamaz.")
else:
    for i in range(1, 2*n):
        if i <= n:
            print(' '*(n-i) + '* '*(i))
        else:
            print(' '*(i-n) + '* '*(2*n-i))
```
